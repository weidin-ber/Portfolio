```python
import pandas as pd
import numpy as np
```

```python
# Load the data as DF
data = pd.read_csv("data/Churn_Modelling.csv")
```

```python
# Check data types
data.dtypes
```

```python
# Drop na values
data.dropna(inplace=True)
```

```python
# Get dummies for Geography an Gender
data = pd.get_dummies(data, columns=["Geography", "Gender"], prefix= "dmy",drop_first = True)
```

```python
data
```

```python
# Convert customerID to index and drop Rownumber and Surname
data=data.set_index('CustomerId')
data = data.drop("RowNumber", axis=1)
data = data.drop("Surname", axis=1)
```
