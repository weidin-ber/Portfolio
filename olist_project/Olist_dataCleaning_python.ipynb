**Clean data for search volume**

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df_dailysales = pd.read_csv('Data/order_payments_sum_review_cleaned.csv')
```

```python
df_dailysales
```

```python
#create an new column only showing the date (split formula)
df_dailysales['date'] = df_dailysales['order_purchase_timestamp'].str.split(' ').str[0]
```

```python
df_dailysales
```

```python
#drop the columns that we don't need
df_dailysales = df_dailysales.drop(columns=['order_purchase_timestamp','order_id','payment_type','review_score'], axis=1)
```

```python
df_dailysales.sort_values(by=['date'])
```

```python
#one row per day. We need to group by date and sum all the payments. 
df_daily_totalsales = df_dailysales.groupby(['date'])[['payment_sum']].sum()
```

```python
#we don't want to have the date as index
df_daily_totalsales['date']=df_daily_totalsales.index
```

```python
#create an new column only showing the year
df_daily_totalsales['year'] = df_daily_totalsales['date'].str.split('-').str[0]


df_daily_totalsales.head()
```

```python
df_daily_totalsales['date']= pd.to_datetime(df_daily_totalsales['date'])
```

```python
df_daily_totalsales["week_number"] = df_daily_totalsales.date.dt.isocalendar().week
```

```python
df_daily_totalsales
```

```python
#we'll clean the data to have it ready to group by
df_weekly_totalsales = df_daily_totalsales.groupby(['year','week_number'])[['payment_sum']].sum()
df_weekly_totalsales
```

```python
#Now we'll download the data from Google trends

df_googletrends_olist = pd.read_csv('Data/googletrends.csv')
```

```python
#we need to clen this df

print(df_googletrends_olist.info())
print(df_googletrends_olist.head())
```

```python
df_googletrends_olist['week']=df_googletrends_olist.index
```

```python
df_googletrends_olist.rename(columns ={"Category: All categories":"interest_over_time"})
```

```python
#clean the first row
df_googletrends_olist.drop('Week',axis=0,inplace=True)
```

```python
df_googletrends_olist.head()
```

```python
df_googletrends_olist['week']= pd.to_datetime(df_googletrends_olist['week'])
```

```python
df_googletrends_olist["week_number"] = df_googletrends_olist.week.dt.isocalendar().week
```

```python
df_googletrends_olist
```

```python
df_googletrends_olist = df_googletrends_olist.rename(columns ={"Category: All categories":"interest_over_time"})
```

```python
df_googletrends_olist = df_googletrends_olist.drop(columns=['week'], axis=1)
```

```python
df_final = pd.concat([df_weekly_totalsales,df_googletrends_olist])
```
