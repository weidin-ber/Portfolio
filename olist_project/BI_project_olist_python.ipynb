```python
#Import the sales csv and the google trends csv
sales_df = pd.read_csv("data/cleaned/order_payments_sum_review_cleaned.csv")
google_df = pd.read_csv("data/multiTimeline.csv")
```

```python
sales_df
```

```python
#split the date column and remove the the time from the date, save the date as new date
sales_df['new_date'] = sales_df['order_purchase_timestamp'].str.split(' ').str[0]
```

```python
#save the year as new column
sales_df["year"] = sales_df['order_purchase_timestamp'].str.split('-').str[0]
```

```python
sales_df
```

```python
#save the new date as datetime datatype
sales_df['new_date'] = pd.to_datetime(sales_df['new_date'])
```

```python
#use the .isocalender function to get the week number of that date
sales_df['week'] = sales_df.new_date.dt.isocalendar().week
```

```python
sales_df
```

```python
google_df
```

```python
#create a new column that uses the index as content
google_df['date'] = google_df.index
```

```python
#create a new column year that only contains the year
google_df["year"] = google_df['date'].str.split('-').str[0]
```

```python
#Drop the two columns in the sales df
sales_df.drop(columns=["order_purchase_timestamp", "new_date"], inplace=True)
```

```python
sales_df
```

```python
#Group sales_df by Year and week to get the payment sum per week
sales_df = sales_df.groupby(['year','week'])[['payment_sum']].sum()
```

```python
#change dtype of year to string
sales_df['year'] = sales_df['year'].astype(str)
```
